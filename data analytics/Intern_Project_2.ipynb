```python
# Mean of numeric columns
print("Mean:\n", df.select_dtypes(include='number').mean())

# Median
print("\nMedian:\n", df.select_dtypes(include='number').median())

# Mode (most frequent values)
print("\nMode:\n", df.mode().iloc[0])

# Standard Deviation
print("\nStandard Deviation:\n", df.select_dtypes(include='number').std())

# Minimum and Maximum values
print("\nMinimum Values:\n", df.select_dtypes(include='number').min())
print("\nMaximum Values:\n", df.select_dtypes(include='number').max())

# Correlation between numeric columns
corr = df.corr(numeric_only=True)
print(corr)
#print("\nCorrelation:\n", df.select_dtypes(include='number').corr())

# Summary Statistics
print("\nSummary:\n", df.describe())

```

```python
# Sort data by Date
df['Date'] = pd.to_datetime(df['Date'])
df_sorted = df.sort_values('Date')

# Check total confirmed cases trend over time
trend = df_sorted.groupby('Date')['Confirmed'].sum()
print(trend.tail(10))  # show last 10 days to see pattern
```
