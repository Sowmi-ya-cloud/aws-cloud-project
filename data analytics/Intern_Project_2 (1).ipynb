```python
# Mean
print("Mean:\n", df.mean(numeric_only=True), "\n")

# Median
print("Median:\n", df.median(numeric_only=True), "\n")

# Mode (most frequent values)
print("Mode:\n", df.mode().iloc[0], "\n")

# Standard Deviation
print("Standard Deviation:\n", df.std(numeric_only=True), "\n")

# Correlation between numeric columns
corr = df.corr(numeric_only=True)
print(corr)

```

```python
# convert to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Sort by date
df_sorted = df.sort_values('Date')

# Check trend of confirmed cases over time
trend = df_sorted.groupby('Date')['Confirmed'].sum()

print("\nConfirmed Cases Trend (Last 10 entries):\n")
print(trend.tail(10))
```
